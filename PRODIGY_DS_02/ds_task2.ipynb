```python
import pandas as pd

# Load the dataset
df = pd.read_csv('test.csv')

# Display the first few rows to understand the dataset
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display missing values
print(missing_values)
```

```python
# Fill missing values in 'Age' with the median age
df['Age'] = df['Age'].fillna(df['Age'].median())

# Safely drop 'Cabin' column
if 'Cabin' in df.columns:
    df.drop(columns='Cabin', inplace=True)

# Fill missing values in 'Embarked' with the most frequent value (mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Convert 'Sex' column to numerical: male = 1, female = 0
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})

# Convert 'Embarked' to numerical values: C = 0, Q = 1, S = 2
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
Exploratory data analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Plot the distribution of Passenger Class
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass')
plt.title('Passenger Class Distribution')
plt.show()

# Plot the distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()

```
